```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"Concrete_Data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
print(df.columns)
print(df.shape)
```

```python
df.drop([0], inplace=True)
df = df.replace(r'\[|\]','',regex=True).astype(float)
df.head()
print(type(df), df.dtypes)
cor = df.corr(numeric_only = False)
cor
```

```python
plt.figure(figsize = (8 ,8))
sns.heatmap(cor , cmap='Blues', annot=True)
plt.show()
```

```python
x = df.iloc[:,:-1].to_numpy()
y = df.iloc[:, -1].to_numpy().reshape(-1, 1)
x.shape , y.shape
print('x' , x)
print('y' , y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, shuffle=True)
print(x_train.shape , x_test.shape)
print(type(x_train))
```

```python
model1 = LinearRegression()
Cement_train =  x_train[:,0].reshape(-1,1)
model1.fit(Cement_train, y_train)
```

```python
weights = model1.coef_
bias = model1.intercept_
print('weights', weights)
print('bias', bias)
```

```python
Cement_test = x_test[:, 0].reshape(-1, 1)
y_hat_test = model1.predict(Cement_test)
```

```python
plt.plot(Cement_train, y_train, 'om', label='cement(train)')
plt.plot(Cement_test, y_test, 'ob', label='true cement')
plt.plot(Cement_test, y_hat_test, 'o-r', label='predict cement')
plt.xlabel("Cement (component 1)(kg in a m^3 mixture)")
plt.ylabel("Concrete compressive strength(MPa, megapascals)")
plt,plt.title("Concrete compressive strength---Cement")
plt.legend()
plt.show()
```

```python
def mse_err(y1, y2):
    return np.mean((y1-y2)**2)
mse_err(y_hat_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
err = mean_squared_error(y_test, y_hat_test)
err
```

```python
model2 = LinearRegression()
two_train = np.stack((x_train[:,0], x_train[:,4]), axis=1)
three_train = np.append(two_train, x_train[:,7].reshape(-1,1), axis=1)
print(three_train)
model2.fit(three_train, y_train)
```

```python
model2.coef_ , model2.intercept_
```

```python
two_test = np.stack((x_test[:,0], x_test[:,4]), axis=1)
three_test = np.append(two_test, x_test[:,7].reshape(-1,1), axis=1)
y_hat_test = model2.predict(three_test)
err = mean_squared_error(y_test, y_hat_test)
print(err)
```
